```python
import uproot
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
file_name = './input/FlashID_nue_v11.root'
uproot.open(file_name)['pandoraFlashEventBuilding'].keys()
```

```python
flashes = uproot.open(file_name)['pandoraFlashEventBuilding'][b'flashes']
events = uproot.open(file_name)['pandoraFlashEventBuilding'][b'events']
slices = uproot.open(file_name)['pandoraFlashEventBuilding'][b'slices']
subruns = uproot.open(file_name)['pandoraFlashEventBuilding'][b'subruns']
meta = uproot.open(file_name)['pandoraFlashEventBuilding'][b'metadata']
forest = [meta, events, flashes, slices, subruns]
```

```python
for tree in forest:
    print(tree.numentries)
    print(tree.keys())
```

```python
print(sum(events.array(b'foundATargetSlice'))/events.numentries)
```

```python
print("Total POT", sum(subruns.array(b'pot')))
print(sum(events.array(b'nSlices')))
```

```python
for key in slices.keys():
    data = slices.array(key)
    print(key, data.dtype)
    if data.dtype!='object':
        if data.dtype=='bool':
            data=data.astype(int)
        plt.hist(data, bins=40)
        plt.show()
```

```python
for key in flashes.keys():
    data = flashes.array(key)
    print(key, data.dtype)
    if data.dtype!='object':
        if data.dtype=='bool':
            data=data.astype(int)
        plt.hist(data, bins=40)
        plt.show()
```

```python
for key in events.keys():
    data = events.array(key)
    print(key, data.dtype)
    if data.dtype!='object':
        if data.dtype=='bool':
            data=data.astype(int)
            
        x_min = min(data)
        if(x_min<-1e4):
            x_min = -max(data)
            
        plt.hist(data, range = (x_min, max(data)), bins=40))
        plt.show()
```
